```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
def read(path,tipe,kelas):
    files = Path(path).glob('*.csv') 
    dfs = [pd.read_csv(f) for f in files]
    ndf=[]
    if tipe=='maxmin' :
        for i in range(len(dfs)):
            x=[]
            maks = max(dfs[i]['S1'])-min(dfs[i]['S1'])
            x.append(maks)
            maks = max(dfs[i]['S2'])-min(dfs[i]['S2'])
            x.append(maks)
            maks = max(dfs[i]['S3'])-min(dfs[i]['S3'])
            x.append(maks)
            maks = max(dfs[i]['S4'])-min(dfs[i]['S4'])
            x.append(maks)
            maks = max(dfs[i]['S5'])-min(dfs[i]['S5'])
            x.append(maks)
            maks = max(dfs[i]['S6'])-min(dfs[i]['S6'])
            x.append(maks)
            maks = max(dfs[i]['S7'])-min(dfs[i]['S7'])
            x.append(maks)
            maks = max(dfs[i]['S8'])-min(dfs[i]['S8'])
            x.append(maks)
            maks = max(dfs[i]['S9'])-min(dfs[i]['S9'])
            x.append(maks)
            maks = max(dfs[i]['S10'])-min(dfs[i]['S10'])
            x.append(maks)
            ndf.append(x)

        df = pd.DataFrame(ndf, columns=["s1", "s2",'s3','s4','s5','s6','s7','s8','s9','s10'])
        df['class']=kelas
        return df
        
    elif tipe=='max' :
        for i in range(len(dfs)):
            x=[]
            maks = max(dfs[i]['S1'])
            x.append(maks)
            maks = max(dfs[i]['S2'])
            x.append(maks)
            maks = max(dfs[i]['S3'])
            x.append(maks)
            maks = max(dfs[i]['S4'])
            x.append(maks)
            maks = max(dfs[i]['S5'])
            x.append(maks)
            maks = max(dfs[i]['S6'])
            x.append(maks)
            maks = max(dfs[i]['S7'])
            x.append(maks)
            maks = max(dfs[i]['S8'])
            x.append(maks)
            maks = max(dfs[i]['S9'])
            x.append(maks)
            maks = max(dfs[i]['S10'])
            x.append(maks)
            ndf.append(x)

        df = pd.DataFrame(ndf, columns=["s1", "s2",'s3','s4','s5','s6','s7','s8','s9','s10'])
        df['class']=kelas
        return df
    else:
        print('masukan yg benar')
        
```
